Firstly, I will encode all the columns that are non-numerical so that we can calculate correlation, and also because the ML model only accepts numerical data. So, encoding is necessary.

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

objects = df.select_dtypes(include = 'object').columns # Here, I am extracting those columns which are of object type.
objects 
```

```python
for i in objects:
    df[i] = le.fit_transform(df[i])

df
```

```python
df.nunique()
# Here, I am checking how many number of unique values each column has. If the number is too high then that column is not a good feature.
# Here, columns like Address have too many unique values. So, we will drop the column.
```

Now, I am assuming that the price column is the one to be predicted, as in most datasets. So, I will be looking at the correlations of each column with respect to the price column. If the value of correlation is too low or too high, then it is not a good feature for ML model.

```python
corr = df.corr()
print(corr['Price'])
```

Here, columns like address has correlation very close to 0 but very high number of unique values. So, it will be dropped. Also, columns like lattitude and longitude do not make a difference for the price column and they have high number of unique values, if we just think about it. So, they will be dropped as well. 

Some columns like Building area has very good correlation but almost 50% of it is empty. So, we will be looking at the Building area column.

```python
df['BuildingArea'].describe()
```

We will replace all the null values of building area with its mean.

```python
df['BuildingArea'].fillna(value=df['BuildingArea'].mean(), inplace=True)
df['BuildingArea'].describe()
```

```python
#Looking at null value count once more
df.isnull().sum()
```

Here, YearBuilt has high number of null values and also its correlation with respect to the price column is relatively low (-0.323617). So, we will also be dropping this column.

As for the null values in the Car column, they are too low of a number as compared to total number of rows (13580). So, we will simply be dropping those rows.

```python
df.dropna(subset=['Car'], inplace=True) # Dropping the rows which have null values in the car column.
df.isnull().sum()
```

Now, I will be dropping all the columns of which I have mentioned the name above.

```python
df.drop(columns=['Address', 'Lattitude', 'Longtitude', 'YearBuilt'], axis=1, inplace=True)
df
```
